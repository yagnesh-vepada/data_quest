```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import metrics
data=pd.read_excel("Data Sheet.xlsx")
```

```python
data
```

```python
data = data[data['Road_Surface_Condition'] == 'Wet']
```

```python
print(data)
```

```python
{col: len(data[col].unique()) for col in data.columns if data.dtypes[col]=='object'}
```

```python
def onehot_encode(df,columns,prefixes):
    df=df.copy()
    for column,prefix in zip(columns,prefixes):
        dummies=pd.get_dummies(df[column],prefix=prefix)
        df=pd.concat([df,dummies],axis=1)
        df=df.drop(column,axis=1)
    return df
```

```python
data=data.drop(['Road_Surface_Condition'],axis=1)
```

```python
data
```

```python
y=data.Crash_Severity
```

```python
y
```

```python
data=onehot_encode(
    data,
    columns=['Crash_Severity', 'Gender',
       'Vehicle_Type', 'Road_Type',
       'Alcohol_Consumption', 'Crash_Type', 'Seatbelt_Usage'],
    prefixes=['cs','g','vt','rt','ac','ct','su']

)
```

```python
data
```

```python
x=data.astype(np.float64)
```

```python
x
```

```python
y=data[['cs_Fatal crash',	'cs_Major injury',	'cs_Minor injury']].copy()
```

```python
y
```

```python
x=x.drop(['cs_Fatal crash'	,'cs_Major injury'	,'cs_Minor injury'	],axis=1)
```

```python
x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.15)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
rf.score(x_test,y_test)
```

```python
import tensorflow as tf
```

```python
inputs=tf.keras.Input(shape=(x.shape[1],))
x=tf.keras.layers.Dense(8,activation='relu')(inputs)
x=tf.keras.layers.Dense(8,activation='relu')(x)
outputs=tf.keras.layers.Dense(3,activation='softmax')(x)
model=tf.keras.Model(inputs,outputs)
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
batch_size=10
epochs=20
history=model.fit(
    x_train,
    y_train,
    validation_split=0.2,
    batch_size=batch_size,
    epochs=epochs,
    callbacks=[
        tf.keras.callbacks.ReduceLROnPlateau(),
        tf.keras.callbacks.EarlyStopping(
            monitor='val_loss',
            patience=3,
            restore_best_weights=True
        )
    ]
)
```
